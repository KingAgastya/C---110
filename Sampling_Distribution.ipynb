```python
import plotly.figure_factory as ff
import pandas as pd
import csv

f = pd.read_csv('temp.csv')

fileData = f['temp'].tolist()

fig = ff.create_distplot([fileData], ['Temperatures'], show_hist = False)
fig.show()
```

```python
import statistics as st
import random

means = []

def funct(n):
  numbers = []
  for i in range(0, n):
      random_index = random.randint(0, (len(fileData)-1))
      numbers.append(fileData[random_index])

  mean = st.mean(numbers)
  return mean


for j in range(0, 1000):
  meanData = funct(100)
  means.append(meanData)

print(means)

fig = ff.create_distplot([means], ['Temperatures'], show_hist = False)
fig.show()

average_1 = st.mean(fileData)
average_2 = st.mean(means)

median_1 = st.median(fileData)
median_2 = st.median(means)

mode_1 = st.mean(fileData)
mode_2 = st.mean(means)

deviation_1 = st.stdev(fileData)
deviation_2 = st.stdev(means)

print(average_1, median_1, mode_1, deviation_1)
print(average_2, median_2, mode_2, deviation_2)
```
